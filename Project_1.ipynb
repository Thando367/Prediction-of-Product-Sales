```python
# fixing inconsistancies
# for Item_Fat_Content
def fat_content(value):
    value = value.lower()
    if 'low fat' in value:
        return 'Low Fat'
    elif 'reg' in value:
        return 'Regular'
    elif 'lf' in value:
        return 'Low Fat'
    else:
        return value

df_sales['Item_Fat_Content'] = df_sales['Item_Fat_Content'].apply(fat_content)
```

```python
# fixing inconsistancies for Outlet_Type
def outlet_type(value):
  value = value.lower()
  if 'supermarket type1' in value:
    return 'Supermarket Type1'
  elif 'supermarket type2' in value:
    return 'Supermarket Type2'
  elif 'supermarket type3' in value:
    return 'Supermarket Type3'
  elif 'grocery store' in value:
    return 'Grocery Store'
  else:
        return value

# applying function
df_sales['Outlet_Type'] = df_sales['Outlet_Type'].apply(outlet_type)
df_sales
```

```python
# define feature X and exclude name column
target = 'Item_Outlet_Sales'
X = df_sales.drop(columns=[target]).copy()

# define target y as Item_Outlet_Sales
y = df_sales[target].copy()
```

```python
# performing the train test split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

<h1>Create a preprocessing object to prepare the dataset for Machine Learning

<h3> 1. Numerical

```python
# creating a numerical data selector
num_cols = X_train.select_dtypes('number').columns
num_cols
```

```python
#instantiate preprocessors
miss_imputer = SimpleImputer(strategy='median')
scaler = StandardScaler()

#creating pipeline
num_pipeline = make_pipeline(miss_imputer, scaler)


# Making a numeric tuple for ColumnTransformer
num_tuple = ('numeric', num_pipeline, num_cols)
num_tuple
```

<h4> 2. Categoric

```python
#from pandas.core.dtypes.dtypes import Ordered

# Categorical - column names
cat_cols = X_train.select_dtypes('object').columns
cat_cols
```

```python
# Instantiate the preprocessors
imputer = SimpleImputer(strategy='constant', fill_value='NA')
one_encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore')

# create a pipeline with imputer and onehotencoder
cat_pipeline = make_pipeline(imputer, one_encoder)

# Making a ohe_tuple for ColumnTransformer
cat_tuple = ('categorical', cat_pipeline, cat_cols)
cat_tuple
```

```python
# making a column transformer
preprocessor = ColumnTransformer([num_tuple, cat_tuple],
                                 verbose_feature_names_out=False)
preprocessor
```

```python
# Fit the preprocessor on X_train data
preprocessor.fit(X_train)
```

```python
# Transforming the training and test data
X_train_tf = preprocessor.transform(X_train)
X_test_tf = preprocessor.transform(X_test)
```

```python
import numpy as np

print("Shape:", X_train_tf.shape)
print("Data Type:", X_train_tf.dtype)
```

```python
# function
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
def regression_metrics(reg,X_test,y_test, label='Test Data'):

  # Get predictions
  y_pred = reg.predict(X_test)

  # Get metrics
  mae = mean_absolute_error(y_test, y_pred)
  mse = mean_squared_error(y_test, y_pred)
  rmse = mean_squared_error(y_test, y_pred, squared=False)
  r_squared = r2_score(y_test, y_pred)

  # Print Result with label
  header = "---"*20
  print(header,f"Regresion Metrics: {label}",header, sep='\n')
  print(f"MAE = {mae:,.2f}")
  print(f"MSE = {mse:,.2f}")
  print(f"RMSE = {rmse:,.2f}")
  print(f"R^2 = {r_squared:,.2f}")



def evaluate_regression(reg, X_train, y_train, X_test, y_test):

  # Get results for training data
  regression_metrics(reg, X_train, y_train, label='Training Data' )

  # Get results for test data
  regression_metrics(reg, X_test, y_test, label='Test Data' )
```

# Modeling

### Linear Modeling

```python
from sklearn.linear_model import LinearRegression
```

```python
# creating linear regression
linear_reg = LinearRegression()
linear_reg
```

## Random Forest

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
# creating Random Forest Model
random_forest = RandomForestRegressor(random_state=42)
# creating pipeline
rf_pipeline = make_pipeline(preprocessor, random_forest)
# fitting the model
rf_pipeline.fit(X_train, y_train);
# evaluate
evaluate_regression(rf_pipeline, X_train, y_train, X_test, y_test)
```



*   R^2 (R-Squared) :
        *   Training data - 0.94
        *   Test data - 0.55

*   The training R-Squared value is greater than the tes R-Squared value. The is a huge gap between the training and test R-squared values, Random Forest might be overfitting the training data.

*   To improve the model's performance we need to do fine-tuning.





## GridSearchCV (Tuning)

```python
from sklearn.model_selection import train_test_split, GridSearchCV
```

```python
# Options for tuning
rf_pipeline.get_params()
```

```python
# Define param grid with options to try
params = {'randomforestregressor__max_depth': [None,10,15,20],
          'randomforestregressor__n_estimators':[10,100,150,200],
          'randomforestregressor__min_samples_leaf':[2,3,4],
          'randomforestregressor__max_features':['sqrt','log2',None],
          'randomforestregressor__oob_score':[True,False],
          }
```

```python
# instantiate the GridSearch
grid_s = GridSearchCV(rf_pipeline, params, n_jobs=-1, cv = 3, verbose = 1)

# Fitting the GridSearch on the training data
grid_s.fit(X_train, y_train)
```

```python
# getting the best params
grid_s.best_params_
```

```python
# Defining best model
rf_best = grid_s.best_estimator_
```

```python
# refit
evaluate_regression(rf_best, X_train, y_train, X_test, y_test)
```


Tuned Random Forest:

*   R^2 (R-Squared) :
        *   Training data - 0.70
        *   Test data - 0.59

*   Compared to the default model Tuned Random Forest have a better performance.

*   The gap between trsining and test R-squared values has been reduced.
